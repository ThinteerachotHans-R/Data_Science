## **PCA with 95% Variance + Logistic regression**

```
from sklearn.decomposition import PCA

pca = PCA(0.95,random_state=RS)  #95 percent of variance
X_pca = pca.fit_transform(X)
X_pca.shape
```

```
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=RS)
```

## **PCA With 2 Modified Parameters**

```
rpca = PCA(n_components=29,svd_solver='randomized',random_state=RS,iterated_power=6) #define the number of components is equal two
X_rpca = rpca.fit_transform(X_scaled)
X_rpca.shape
```

```
X_train_rpca, X_test_rpca, y_train, y_test = train_test_split(X_rpca, y, test_size=0.2, random_state=RS)
log_re = LogisticRegression(max_iter=1000,random_state = RS)
log_re.fit(X_train_rpca, y_train)
mod_score = log_re.score(X_test_rpca, y_test)
print(mod_score)
```

```
accuracy = []
for i in range(1,11):
  rpca = PCA(n_components=29,svd_solver='randomized',random_state=RS,iterated_power=i) #define the number of components is equal two
  X_rpca = rpca.fit_transform(X_scaled)
  X_rpca.shape

  X_train_rpca, X_test_rpca, y_train, y_test = train_test_split(X_rpca, y, test_size=0.2, random_state=RS)


  log_re = LogisticRegression(max_iter=1000,random_state = RS)
  log_re.fit(X_train_rpca, y_train)
  mod_score = log_re.score(X_test_rpca, y_test)
  print(f'Accuracy for {i} iterated power =  {mod_score}')
  accuracy.append(mod_score)

```

```
import matplotlib.pyplot as plt
import numpy as np

# Assuming all_score, nf_score, svp_score, and accuracy are defined

# Use the accuracy for iterated_power=1 from the accuracy list
pca_iterated_power_1_score = accuracy[0]

scores = [all_score, nf_score, pca_iterated_power_1_score]
labels = ['All Features', 'PCA (95% Variance)', 'PCA (Iterated Power 1)']

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, scores, color=['blue', 'orange', 'green', 'purple'])

plt.ylabel('Accuracy Score')
plt.title('Comparison of Logistic Regression Accuracy Scores')
plt.ylim(0, 1) # Assuming scores are between 0 and 1
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.4f}', va='bottom', ha='center') # Format to 4 decimal places


plt.tight_layout()
plt.show()
```

## IRIS - Dataset

```
import pandas as pd
```

```
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
df.head()
```

```
df.sample(10)
```

```
df.describe()
```

```
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
df.hist(layout=(1, len(features)), figsize=(3 * len(features), 3.5));
```

```
df.corr(numeric_only=True)
```

```
import numpy as np
import seaborn as sns

dcorr = df[features].corr()

mask = np.zeros_like(dcorr)
#mask shape
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(dcorr, cmap=sns.diverging_palette(10, 145, n=100),
            vmin=-1, vmax=1, center=0, linewidths=1, annot=True, mask=mask, ax=ax);
```

```
sns.pairplot(df, vars=features,
             hue='target');
```

```
from sklearn.preprocessing import StandardScaler
# Separating out the features
x = df.loc[:, features].values
# Separating out the target
y = df.loc[:,['target']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```
x_df = pd.DataFrame(x)
x_df.head()
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf.head()
```

```
finalDf = pd.concat([principalDf, df[['target']]], axis = 1)
```

```
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

### **Logistic Regression + PCA**

```
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

accuracy_scores = []
n_components_range = range(1, 8)

for i in n_components_range:
  pca = PCA(n_components=i, random_state=42)
  # Use x_scaled for PCA as it's the standardized data
  X_pca = pca.fit_transform(x)

  X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y.flatten(), test_size=0.2, random_state=42)

  model = LogisticRegression(max_iter=1000)
  model.fit(X_train_pca, y_train_pca)
  a_score = model.score(X_test_pca, y_test_pca)
  accuracy_scores.append(a_score)
  print(f"Number of components: {i}, Accuracy: {a_score:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(list(n_components_range), accuracy_scores, marker='o', linestyle='-')
plt.xlabel('Number of Principal Components')
plt.ylabel('Accuracy Score')
plt.title('Logistic Regression Accuracy vs. Number of PCA Components')
plt.xticks(list(n_components_range))
plt.grid(True)
plt.show()
```

```
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(pca.n_components_)], index=column[:-1])
sns.heatmap(loadings, cmap='coolwarm', center=0, annot=True)
plt.title('PCA Loadings Heatmap')
plt.show()
```
